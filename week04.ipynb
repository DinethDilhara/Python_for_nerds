# <font color="green">**While Loop**</font>

A condition-controlled loop executes statements if a `condition is true`.





```python
i=0 #intialize control
while i<3:#i<3 -> condition
    print('i =',i)
    i+=1#counter update
print("Done")
```

# <font color="green">**For Loop**</font>


A for-loop is used to step through a sequence.

Also known as a `count controlled` loop

The `for` and `in` are key words that are inbuiltcommands.

        Syntax:
        for<name of loop control> in <something that can be iterated>:
                          body

## Range Function

The range() function can be used to create a sequence of numbers.
 The structure of the function is **range(start, stop, step)**.
-     o start (optional): starting number of sequence. he default is 0.
-     o stop: generate sequence up to, but not including the specified number.
-     o step (optional): number specifies the sequence incrementation. The default is 1.


```python
#start value =0
# end value =5
#setp =1
for i in range (5):  
    print(i)
```

```python
#start value =10
# end value =5
#setp =-1
for i in range (10,5,-1):  
    print(i)
```

# </font><font color="green">**Control flow statements**</font>

continue, break, and pass are `control flow statements` used within repetitive statements

| Keyword   | Description                                             |
|-----------|---------------------------------------------------------|
| `pass`    | Placeholder statement; **does nothing** and moves to next. |
| `continue`| **Skips** the rest of the code inside a loop and goes to the next iteration. |
| `break`   | **Exits** the loop prematurely.                               |



```python
for i in range(5):
    if i == 2:
        pass  # Placeholder, does nothing
    else:
        print(i)
```

```python
for i in range(5):
    if i == 2:
        continue  # Skips the rest of the loop and goes to the next iteration
    else:
        print(i)
```

```python
for i in range(5):
    if i == 2:
        break  # Exits the loop prematurely
    else:
        print(i)
```

# </font><font color="green">**Nested Repetitive Statements**</font>



        syntax:      outer loop(run n times):
                         inner loop(run m times):
                              body(run n*m times)


```python
for i in range(3):
   for j in range(2):
         print(i, j)
print("done")
```
