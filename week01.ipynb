# <u>**Python identifiers**</u>

- Python is case-sensitive.
- An identifier is a name used to identify a variable, function, class, module, or object.
- Identifiers can only **start with an alphabetical character or underscore (_).**
- After the first letter, any alphabetical character or number can be used.
- There is no limit to the number of characters for identifier names.
- Reserved words cannot be used as identifiers.


# <u>**Variables**</u>

- Variables follow all identifier rules
- To assume some value to the variable we use (=)
- Variable names are **case sensitive**
- <u>**cannot contain spaces First character must be a letter or an underscore**</u>
- cannot be a Python keyword

# <u>**Strings**</u>

    ->a sequence of characters

    ->In Python, write a string using ““ double quotes or ‘ ‘  single quotes.



<u><font color="green">**01.String Concatenation.**</font></u>

**-->(+) The operator performs string concatenation.**

**-->(*) the operator performs repetition on a string.**


```python
Frist_name = "Dineth"
second_name = "Dilhara"
age = 16

print( Frist_name + second_name)
print( Frist_name *3)
```

<u><font color="green">**02.print() with string concatenation** .</font></u>

**-->To string concatenation requires strings. If there is any other data type it can be converted to string using str ().**

**-->(+) operator does not add extra space between strings but (,) does.**

```python
print(Frist_name + "is" + str(age) +"years old.")
print(Frist_name,"is",str(age),"years old.")
```

<u><font color="green">**03.String escape sequence** .</font></u>

 In Python, we can use, 
-       ->  Single quotes in double quotes #1
-       ->  Double quotes in single quotes #2
-       ->  Escape sequence #3
-       -> use ''' triple quotes to multiple lines #4

<u>Common escape sequence</u>

| Sequence | Meaning |
|----------|----------|
| \\ | Literal back|
| \' | single quote |
| \" | double quote|
| \n | New line |
| \t | Tab|



```python
print("#1 and #2")
print("i'm Dineth")#1
print('dineth said "hello!" ')#2
print("----")
print("#3")
print("do you want tea\\coffee")
print('I\'m Dineth')
print("I\"m Dineth")
print()
print("#New Line")
print("my name is dineth \nwhat is yours?")
print()
print("#Tab")
print("my name is dineth \twhat is yours?")
print("----")
print("#4")
print('''my name is dineth 
what is yours?''')
```
