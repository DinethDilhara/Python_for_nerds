# <font color="green">**Boolean**</font>

Booleans represent one of two values: <font color="red">True</font> or <font color="red">False.</font>


```python
#using comparison operation find statement is whether it is true or false 
print(10==5)
print(10<5)
print(10>5)
```

### <u>Boolean Variable </u>

A Boolean variable is a type of variable that can only take on one of two values: True or False.

Boolean variables are often used to represent binary conditions in programming, where something is either true or false


### <u>Boolean expression</u>

Boolean expressions are conditions that evaluate to either True or False.





```python
#Boolean Variable
am_i_dineth = True

print(am_i_dineth)
```

```python
# Boolean expression
x = 5
y = 10
is_greater = (x > y)
print(is_greater)
```

### <u>Boolean Operators</u>

Python provides three main boolean operations: **and**, **or**, and **not**.
#----------------------------------------------------------------------------------------#
### AND 

Evaluate two expressions. Evaluate to **True if both of the two values are True.**

|   A   |   B   | Result |
|-------|-------|--------|
| True  | False | False  |
| True  | True  | True   |
| False | False | False  |
| False | True  | False  |
| True  | False | False  |





```python
x = 50
y = 10

print (x==50 and y==10)
print (x==50 and y==1)
print (x==5 and y==10)
print (x==5 and y==1)
```

#----------------------------------------------------------------------------------------#
### OR

Evaluate two expressions. Evaluate to **True if either of the two values is True.**
|   A   |   B   | Result |
|-------|-------|--------|
| True  | False | True   |
| True  | True  | True   |
| False | False | False  |
| False | True  | True   |



```python
x = 50
y = 10

print (x==50 or y==10)
print (x==50 or y==1)
print (x==5 or y==10)
print (x==5 or y==1)
```

#----------------------------------------------------------------------------------------#
### NOT

not operator evaluates to the **opposite Boolean value – Inverts a condition.**

|   A   | Result |
|-------|--------|
| True  | False  |
| True  | False  |
| False | True   |
| False | True   |
| True  | False  |

```python
x=50

print(not x==50)
print(not x==1)
```

# <font color="green">**Selection Statement**</font>

when the program has to use only **execute code under certain circumstances.**

We use mathematical logic with if, Elif, or  keywords with the condition.

-   - > if   : keyword with **if condition true do this or that**
-   - > elif : keyword for **if the previous condition is not true, then do this**
-   - > else : keyword catches anything that **isn't caught by the above conditions do this**

| Selection Type        | Description                                        | Main Difference                                      |
|------------------------|----------------------------------------------------|-------------------------------------------------------|
| `if`                   | Executes a block of code if a condition is true.   | Basic conditional statement.                           |
| `if-else`              | Executes one block if a condition is true, and another if false. | Handles two possible outcomes.                         |
| `if-elif`              | Allows testing multiple conditions sequentially.   | Conditions are checked one by one until a true one is found. |
| `if-elif-else`         | Extends `if-elif` to include an else block.         | Handles multiple conditions with a default option.    |
| If Ladder              | A series of `if` and `elif` statements sequentially. | Checks conditions one by one, and executes the first true block. |
| Nested If              | Includes one or more `if` statements inside another `if` or `else` block. | Allows for conditional statements within conditional statements. ||
| Short Hand `if`    | Executes a statement if a condition is true.        | Concise form of the basic `if` statement.               |
| Short Hand `if-else`| Executes one statement if true, another if false.   | Concise form of the `if-else` statement.               |


#-----------------------------------------------------------------------------#

## if 

The basic `if` statement checks a single condition.


```python
a = 50

if a>49:
    print("a is greater than 49")
```

#-----------------------------------------------------------------------------#
## if - else

`if-else` handles two possible outcomes based on a condition

```python
a = 50
b = 100

if a>b:
    print("a is greater than b")
else:
    print("b is greater tha a")
```

#-----------------------------------------------------------------------------#
## if - elif

`if-elif` checks multiple conditions sequentially, executing the first true block.



```python
a = 7

if a > 10:
    print("a is greater than 10")
elif a > 5:
    print("a is greater than 5 but not greater than 10")
```

#-----------------------------------------------------------------------------#
## if - elif - else

`if-elif-else` extends `if-elif` to include a default option if none of the conditions are true.

```python
a = -2

if a > 5:
    print("a is greater than 5")
elif a > 0:
    print("a is positive but not greater than 5")
else:
    print("a is non-positive")
```

#-----------------------------------------------------------------------------#
## if - ladder

If Ladder is a series of `if` and `elif` statements where each condition is checked sequentially.


```python
score = 59

if score >= 75:
    print("A grade")
elif score >= 65:
    print("B grade")
elif score >= 50:
    print("C grade")
else:
    print("Fail")
```

#-----------------------------------------------------------------------------#

## Nested - if

Nested If allows for conditional statements within conditional statements, creating a hierarchy.

```python
b = 8

if b > 5:
    print("b is greater than 5")
    if b % 2 == 0:
        print("b is also even")
    else:
        print("b is odd")
```

#-----------------------------------------------------------------------------#

## Shorthand if 

Shorthand `if-else` provides a shorter syntax for simple conditional expressions.

```python
a = 2
b = 10

if a < b: print("b is greater than a")
```

#-----------------------------------------------------------------------------#

## Shorthand if  -else

Shorthand `if-elif` offers a compact syntax for concisely handling sequential conditions.

```python
a = 30
b = 33
print("b is greater than a") if a < b else print(" a is greater than b") 
```
