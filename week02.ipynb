## <u>02. Runtime Error</u>

These errors  appear when the **program has started running** also call as <u>exceptions</u>

ex:  NameError , TypeError , ValueError , ZerodivisionError

**------------------------------------------------------------------------------------------**

### <u>**NameError**</u>

NameError:  means you have used a **variable before it has a value**

```python
#NameError (here we have initialize num1 to num2 before num2 have defined)
num1=num2
num2=10
print(num1+num2)
```

```python
#Exception for NameError
try:
    num1=num2
    num2=10
    print(num1+num2)
except NameError:
    print("Name Error")
```

**------------------------------------------------------------------------------------------**

### <u>**TypeError**</u>

**TypeErro**r: often when an expression tries to combine two types that are not compatible 

<u>e.g., integer and string </u>

Can use **type()** to check what type of variable is

```python
#TypeError and check data type
name = "dineth"
age = 10
print(type(name))
print(type(age))
print(name + age )
```

```python
#Exception for TypeError
try:
    name = "dineth"
    age = 10
    print(name + age )
except:
    print("TypeError" )
```

**------------------------------------------------------------------------------------------**
### <u>**ValueError**</u>

ValueError: when an operation or function receives an argument that has the **right type but an inappropriate value**

ex: ask  the user for a number and instead they enter a character t

can use **data conversion** str(),int() etc.. or **try exception** to fix ValueError



```python
#valueerror in here abc is to int() 
name = int("dilhara")
print(name)
```

```python
#Exception for ValueError
try:
    name = int("dilhara")
    print(name)
except:
    print("value Error")
```

**------------------------------------------------------------------------------------------**
### <u>**ZeroDivisionError**</u>

 **ZeroDivisionE**rror**: with Python it is not possible to divide numbers by zero. 

 â€¢	If you attempt to divide by 0, python will throw a ZeroDivisionError

 To fix this add **"if" statement to ignore if value is zero** or **try except**

```python
a = 0
b =100
c = b/a
d = a/b
print(c,d)
```

```python
#Exception for ZeroDivisionError
try:
    a = 0
    b =100
    c = b/a
    d = a/b
    print(c,d)
except ZeroDivisionError:
    print("ZeroDivision Error")
```
